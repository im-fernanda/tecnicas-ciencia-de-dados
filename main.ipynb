```python
import pandas as pd
```

```python
filial_a = pd.read_csv("filial_a.csv")
filial_b = pd.read_csv("filial_b.csv")
filial_c = pd.read_csv("filial_c.csv")
```

```python
filial_a['Data da Venda'] = pd.to_datetime(filial_a['Data da Venda'], format='%Y/%m/%d').dt.strftime('%d/%m/%Y')
filial_a
```

```python
taxa_cambio_dollar = 5.65
filial_b['Preço Unitário'] = filial_b['Preço Unitário'].replace({'R\\$ ': ''}, regex=True).astype(float)
filial_b['Preço Unitário'] = filial_b['Preço Unitário'] / taxa_cambio_dollar

filial_b = filial_b.rename(columns={'Preço Unitário': 'Preço (USD)', 'Data': 'Data da Venda', 'Qtd. Vendida':'Quantidade'})	
filial_b
```

```python
filial_c['Data de Venda'] = filial_c['Data de Venda'].replace({'-': '/'}, regex=True)
filial_c['Data de Venda'] = pd.to_datetime(filial_c['Data de Venda'], format='%Y/%m/%d').dt.strftime('%d/%m/%Y')

taxa_cambio_euro = 6.16
filial_c['Valor Unitário (EUR)'] = filial_c['Valor Unitário (EUR)'] / taxa_cambio_euro
filial_c = filial_c.rename(columns={'Valor Unitário (EUR)': 'Preço (USD)', 'Representante' : 'Vendedor', 'Código do Produto': 'Produto','Data de Venda' : 'Data da Venda', 'Tipo': 'Categoria'})	

filial_c
```

```python
filial_a['Filial'] = 'Filial A'
filial_b['Filial'] = 'Filial B'
filial_c['Filial'] = 'Filial C'
```

```python
filial = pd.concat([filial_a, filial_b, filial_c])
filial['Data da Venda'] = pd.to_datetime(filial['Data da Venda'], format='%d/%m/%Y')

filial['Dia da Venda'] = filial['Data da Venda'].dt.day
filial['Mês da Venda'] = filial['Data da Venda'].dt.month
filial['Ano da Venda'] = filial['Data da Venda'].dt.year
filial
```

```python
filial.to_csv('./filial.csv')
```

```python
tabela_pivot = pd.pivot_table(
    filial,
    values='Quantidade',  # Coluna a ser agregada
    index=['Ano da Venda', 'Mês da Venda', 'Dia da Venda'],  # Dimensões de data
    columns='Categoria',  # Dimensão de tipo de produto
    aggfunc='sum',  # Função de agregação
    fill_value=0  # Preencher valores ausentes com 0
)
tabela_pivot
```

```python
import plotly.io as pio
import plotly.express as px
pio.renderers.default = 'iframe_connected'
```

```python
tabela_long = tabela_pivot.reset_index()
tabela_long = tabela_long.melt(id_vars=['Ano da Venda', 'Mês da Venda'], 
                                var_name='Tipo do Produto', 
                                value_name='Quantidade Vendida')

# Criar uma tabela pivot com Ano e Mês como índices
tabela_cubo = tabela_long.pivot_table(index=['Ano da Venda', 'Mês da Venda'], 
                                       columns='Tipo do Produto', 
                                       values='Quantidade Vendida', 
                                       fill_value=0)
```

```python

# G uma visualização usando o imshow
sobre o cubo considerando somente a data e o tipo do produto como
dimensões. 
fig = px.imshow(
    tabela_cubo.values,
    labels=dict(x="Tipo do Produto", y="Data (Ano-Mês)", color="Quantidade Vendida"),
    x=tabela_cubo.columns,
    y=[f"{ano}-{mes:02d}" for ano, mes in tabela_cubo.index],
    title="Quantidade de Produtos Vendidos por Tipo e Data",
    color_continuous_scale='Viridis'
)

# Exibir o gráfico
fig.show()
```
